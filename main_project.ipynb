```python
# car_rental.py

from datetime import datetime, timedelta

class CarRental:
    def __init__(self, total_cars):
        self.total_cars = total_cars
        self.available_cars = total_cars
        self.rented_cars = 0
        self.rental_records = {}

    def display_available_cars(self):
        print(f"Available cars: {self.available_cars}")

    def rent_car(self, num_cars, rental_period, rental_rate):
        if num_cars > 0 and num_cars <= self.available_cars:
            current_time = datetime.now()
            self.rental_records[current_time] = {"num_cars": num_cars, "rental_period": rental_period}
            self.available_cars -= num_cars
            self.rented_cars += num_cars
            return current_time
        else:
            return None

    def return_car(self, rent_time, rental_period, num_cars, rental_rate):
        if rent_time in self.rental_records:
            return_time = datetime.now()
            rental_duration = return_time - rent_time
            if rental_period == "hourly":
                bill = rental_duration.total_seconds() // 3600 * rental_rate * num_cars
            elif rental_period == "daily":
                bill = rental_duration.total_seconds() // 86400 * rental_rate * num_cars
            elif rental_period == "weekly":
                bill = rental_duration.total_seconds() // 604800 * rental_rate * num_cars
            else:
                return None

            del self.rental_records[rent_time]
            self.available_cars += num_cars
            self.rented_cars -= num_cars
            return bill
        else:
            return None

def main():
    total_cars = 10
    car_rental_system = CarRental(total_cars)

    while True:
        print("\n1. Display available cars")
        print("2. Rent a car")
        print("3. Return a car")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            car_rental_system.display_available_cars()
        elif choice == "2":
            num_cars = int(input("Enter the number of cars to rent: "))
            rental_period = input("Enter rental period (hourly/daily/weekly): ")
            rental_rate = int(input("Enter rental rate per unit: "))
            rent_time = car_rental_system.rent_car(num_cars, rental_period, rental_rate)
            if rent_time:
                print(f"{num_cars} cars rented successfully at {rent_time}")
            else:
                print("Invalid input or insufficient available cars.")
        elif choice == "3":
            rent_time = input("Enter the rental time (in the format 'YYYY-MM-DD HH:MM:SS'): ")
            rent_time = datetime.strptime(rent_time, "%Y-%m-%d %H:%M:%S")
            rental_period = input("Enter rental period (hourly/daily/weekly): ")
            num_cars = int(input("Enter the number of cars to return: "))
            rental_rate = int(input("Enter rental rate per unit: "))
            bill = car_rental_system.return_car(rent_time, rental_period, num_cars, rental_rate)
            if bill is not None:
                print(f"Total bill: ${bill}")
            else:
                print("Invalid input or rental record not found.")
        elif choice == "4":
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
